```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip");
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker");
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="smoker"
);
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time",
);
```

```python
sns.relplot(
    data=tips, x="total_bill", y="tip", hue="size",
);
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size");
```

```python
sns.scatterplot(
    data=tips, x="total_bill", y="tip",
    size="size", sizes=(15, 200)
);
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(data=fmri, x="timepoint", y="signal");
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", hue="event",
);
```

```python
sns.lineplot(
    data=fmri,
    x="timepoint", y="signal",
    hue="region", style="event",
);
```

```python
sns.lineplot(
    data=fmri,
    x="timepoint", y="signal", hue="region", style="event",
    dashes=False, markers=True,
);
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", col="time",
);
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
);
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
sns.histplot(penguins, x="flipper_length_mm")
plt.show()
```

```python
sns.histplot(penguins, x="flipper_length_mm", binwidth=3)
plt.show()
```

```python
sns.histplot(penguins, x="flipper_length_mm", hue="species")
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill")
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", shade=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", shade=True)
plt.show()
```

```python
sns.histplot(penguins, x="flipper_length_mm", hue="species", kde=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x="total_bill", hue="time", cumulative=True)
plt.show()
```
